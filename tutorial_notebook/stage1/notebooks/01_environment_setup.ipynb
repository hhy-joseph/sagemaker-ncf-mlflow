## 3. Environment Validation

Let's verify our setup by importing key packages and checking versions:

````python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import xgboost as xgb

# Print versions
print(f"Python version: {sys.version}")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")
print(f"XGBoost version: {xgb.__version__}")
````

## 4. Final Validation

Let's verify our setup by loading and displaying some sample data:

````python
# Test data loading and visualization
try:
    # Load sample data
    df = pd.read_csv('../../data/googl_data_2020_2025.csv')
    
    # Simple visualization
    plt.figure(figsize=(12, 6))
    plt.plot(df['Close'])
    plt.title('Google Stock Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()
    
    print("Environment setup complete and working correctly!")
except Exception as e:
    print(f"Error during validation: {str(e)}")
````
